```python
from __future__ import division
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# strike of 5 
#inital stock of 7   
def odefunc(u, t):
    dudt = np.zeros(x0.shape)
    #boundary conditions
    dudt[0] = u[0]
    # termial condition for the mesh grid
    dudt[-1] = (7 - 5*np.exp(-k*t))/5 

    #step size
    h = b/(N-1)

    for i in range(1, N-1):
        dudt[i] = (u[i + 1] - 2*u[i] + u[i - 1]) / h**2
        
    return dudt


N = 40
r = 0.065;
sigma = 0.8;
k = r/sigma**2
a = np.log(2/5)
b = np.log(7/5)
t0 = 0
tf = 5

x = np.linspace(a, b, N);
# modified inital condition for pure heat equation 
#without the first derivative and u term removed by choosing
#form of a & b to reduce the lower order spatial deriv

x0 = np.exp((k+1)*x/2) - np.exp((k-1)*x/2)
x0[x0 < 0] = 0

tspan = np.linspace(t0, tf, 20);
sol = odeint(odefunc, x0, tspan)


#%% Make a 3d figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

SX, ST = np.meshgrid(x, tspan)
ax.plot_surface(SX, ST, sol, cmap='YlGnBu_r',
                rstride=1, cstride=1)
ax.set_xlabel('moneyness')
ax.set_ylabel('tao')
ax.set_zlabel('u')
ax.view_init(elev=30, azim=-111)
plt.savefig('pde-transient-heat-3d.png')
plt.show() 
```
